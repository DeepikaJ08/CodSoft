# 1. Read Data

```python
import pandas as pd
df = pd.read_csv('IMDb Movies India.csv', encoding='latin-1')
df.head()
```

### Handling missing values

#### For this dataset, we will have to dropp the missing values

```python
# check missing value count
df.isna().sum()
```

```python
# convert the missing value count to % to check the proportion
na_per1 = round((df.isna().sum() * 100 / df.count()),1)
na_per1
```

```python
# drop missing values in df
df.dropna(axis=0, inplace=True)
df.head()
```

```python
# verfiy the above action
na_per2 = round((df.isna().sum() * 100 / df.count()),1)
na_per2
```

### Converting datatypes 

#### per the df info, we can see that the data type for columns - Year, Duration and Votes is 'object'. We need these column as numeric (int/float) data type for our analysis

```python
df.info()
```

```python
df.head()
```

```python
# Year Column - removing the round brackets and converting to integer data type

df["Year"]=df['Year'].str[1:-1].astype(int)
```

```python
# Duration Column - removing the suffix ' min' and converting to integer data type

df['Duration']=df['Duration'].str.replace(' min','').astype(float)
```

```python
# Votes Column - Observed a value '$5.16M' in Votes column which should be replaced with '5160000'

df['Votes'] = df['Votes'].replace('$5.16M', '5160000')
```

```python
# Votes Column - Remove ',' and convert the column data type to integer

df['Votes'] = df['Votes'].str.replace(',','').astype(int) 
```

#### verify the df info after data type conversion

```python
df.info()
```

# 3. Separate Independent (X) and Target (Y) features 

```python
# Separate Independent (X) and Target (Y) features
X = df.drop(columns=['Name', 'Rating'])
Y = df['Rating']
```

```python
X.head()
```

```python
Y.head()
```

## 4. Create Feature Selection Pipeline for X

```python
catx = list(X.columns[X.dtypes=='object'])
conx = list(X.columns[X.dtypes!='object'])
```

```python
catx
```

```python
conx
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe1 = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())])
```

```python
cat_pipe1 = Pipeline(steps=[('impute', SimpleImputer(strategy='most_frequent')),
                           ('ordinal', OrdinalEncoder())])
```

```python
pre = ColumnTransformer([('num', num_pipe1, conx),
                         ('cat', cat_pipe1, catx)]).set_output(transform='pandas')
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

```python
# Feature Selection
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
modelf = LinearRegression()
sel = SequentialFeatureSelector(estimator=modelf, n_features_to_select='auto', direction='forward')
sel.fit_transform(X_pre, Y)
sel_col = sel.get_feature_names_out()
```

```python
sel_col
```

```python
# Getting important columns in a list
imp_col = []
for i in sel_col:
    s = i.split('__')[1]
    imp_col.append(s)
```

```python
imp_col
```

```python
X_sel = X[imp_col]
X_sel.head()
```

## 5. Create a final pipeline for X

```python
cat_sel = list(X_sel.columns[X_sel.dtypes=='object'])
con_sel = list(X_sel.columns[X_sel.dtypes!='object'])
```

```python
cat_sel
```

```python
con_sel
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2 = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                            ('scaler', StandardScaler())])
```

```python
cat_pipe2 = Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent')),
                            ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
```

```python
pre2 = ColumnTransformer([('num', num_pipe2, con_sel),
                          ('cat', cat_pipe2, cat_sel)]).set_output(transform='pandas')
```

```python
X_sel_pre = pre2.fit_transform(X_sel)
X_sel_pre.head()
```

## 6. Apply Train-Test Split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_sel_pre, Y, test_size=0.2, random_state=42)
```

## 7. Model Building and Evaluation

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

# we are not evaluating KNN and SVM as the dataset has >2000 data rows
```

```python
dct = {'Linear' : LinearRegression(),
       'Decision Tree' : DecisionTreeRegressor(),
       'Random Forest' : RandomForestRegressor(),
       'Gradient Boosting' : GradientBoostingRegressor()}
```

```python
dct.items()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score
Train_MSE = []
Test_MSE = []
Train_R2 = []
Test_R2 = []
Train_CV_R2 = []
for name, model in dct.items():
    # Fit the model
    model.fit(xtrain, ytrain)
    # Predict results for Train and Test
    ypred_train = model.predict(xtrain)
    ypred_test = model.predict(xtest)
    # Get MSE, RMSE, MAE, R2 score
    mse_train = mean_squared_error(ytrain, ypred_train)
    mse_test = mean_squared_error(ytest, ypred_test)
    r2_train = r2_score(ytrain, ypred_train)
    r2_test = r2_score(ytest, ypred_test)
    # Get Cross validated scores - r2, mae, mse
    scores_r2 = cross_val_score(model, xtrain, ytrain, cv=5, scoring = 'r2')
    r2_cv = scores_r2.mean()
    # Append above results in list
    Train_MSE.append(mse_train)
    Test_MSE.append(mse_test)
    Train_R2.append(r2_train)
    Test_R2.append(r2_test)
    Train_CV_R2.append(r2_cv)
```

```python
# Create dictionary with the metrics results
res = {'Name':list(dct.keys()),
       'Train_MSE':Train_MSE,
       'Test_MSE':Test_MSE,
       'Train_R2':Train_R2,
       'Test_R2':Test_R2,
       'Train_CV_R2':Train_CV_R2}
```

```python
# Create a data frame with above results
df_res = pd.DataFrame(res)
df_res
```

### Based on above scores, we can say that Random Forest is performing better than other models

```python
best_model = RandomForestRegressor(n_estimators=1,
                                   max_depth=1,
                                   min_samples_split=5,
                                   min_samples_leaf=5,
                                   criterion='squared_error')
best_model.fit(xtrain, ytrain)
```

```python
# Accuracy in Train
best_model.score(xtrain, ytrain)
```

```python
# Accuracy in Test
best_model.score(xtest, ytest)
```
